## Introduction

Wage, expressed also as pay rate, is the primary outcome variable analyzed in this study. I estimated wages based on income class, the interaction between income class and education level, and major occupation category. A final model was built using full subset selection on a selection of predetermined variables.

The purpose of these analyses is to form a cohesive set of indicators of pay rate and identify disparities. The task of conservatively interpreting the results falls to the author, but also to the reader. This research cannot determine causality because it is not longitudinal. It can only serve to inform the audience as to the probable impact of their economic class, level of education, job category, unionization status, etc. on their pay rates.

This study can only hope to find correlations between lower-status indicators (e.g. lower levels of education, major occupation) and lower wages. It can also hope to point to high-status indicators and their correlation to higher wages. Naturally, these differences exist on a continuum between low-status and high-status. The task of deeply analyzing the origins of those disparities is left to future research, but answers can also be found in the historical record, which point to the likely political causes, expressed in terms of government and business policies.

Income class was determined to be one of the most robust indicators of pay rate, alongside major occupation category. As class increases, wages increase substantially. We found education to be an unreliable predictor, typically correlated to increased wages at some of the highest levels of education (e.g., bachelor's degree and above). Once class is controlled for, no impact of any level of education on pay rate was detected.

Major occupation is a stronger predictor, and there are distinct differences between major occupation categories, with a clear spectrum between the best-paid and worst-paid job areas. As major occupation category "increases", pay rate tends to increase. When income class and major occupation are used to predict pay rate together, the influence of major occupation on pay rate decreases, but the majority of major occupation levels continue to explain differences in pay rate. Significant differences were detected between ununionized major occupations and unionized major occupations, with the greatest effect belonging to union members.

### Pay Rate

Pay rate is important in that it represents how much your work is worth; is your work worth $\$3,200$ a month or $20 an hour? The monthly estimate helps us make sense of how much our labor can pay for, and is also robust in terms of comparing it to someone else. The most digestible, however, is pay rate as an hourly wage. Calculating something, even hypothetically (e.g., salaries), as an hourly pay rate, has one key benefit: it reduces the final estimate, which makes rates easier to compare against each other.

This study is primarily concerned with differences in pay rate between different classes of people (based on household income, education, job class, etc.).

Pay rate is calculated as a function of income earned per week divided by hours worked per week:

$$ \text{weekly pay rate} =  \dfrac{income}{hours}$$

Pay rate describes, essentially, how much money an individual makes per hour, regardless of whether they are paid hourly. This allows us to view the overall trend between the categories, and also provides a point of comparison between employees who are paid hourly and those who receive a salary.

Note that there is a very large difference between the pay rates of hourly and salaried employees: salaried employees systematically receive a higher pay rate.

To simplify the research question, hours worked per week and weekly pay is assumed to be constant. These variables are only representative of one point in time: the week before the interviewees were surveyed.

## Hypotheses

### Hypothesis 1: the income class-only hypothesis

Pay rate is a measure of hourly income. Economic class, which this study simplifies into lower, upper, and middle economic class, is determined by adjusted household income. Classification is detemined by the adjusted income of a household relative to the median adjusted household income of the sample.

Economic mobility in the United States it at a historic low. (Citation needed) Research indicates that the quality of a current individual's job is greatly determined by that individual's starting wealth, which can be used to capitalize on networking and other opportunities that require capital investment that is easier to achieve for individuals with higher starting capital. (Citation needed)

Hypothesis 1 posits that the pay rate of working Americans is explained to a great degree by current income class. As income class increases, pay rate increases. Lower-income individuals will have the least pay rate, middle-income individuals will be in the center, and upper-income individuals will have the highest pay rate.

This hypothesis is likely to be true due to the causal relationship between job income and household income. For one-person households, or households where one individual earns a the majority of income, the job income of that individual will largely determine their household, and therefore their income class.

If hypothesis 1 is true, it indicates that the sample population's pay rate is strongly related to the income of the household they belonged to last year (since it is a gross figure, it can only be collected for the prior year).

$$f(x) = \beta_{0} + \beta_{middle-income}X_{middle-income} + \beta_{upper-income}X_{upper-income} +\epsilon$$

Where $\beta_{0}+\epsilon$ is the baseline term and represents the coefficient estimate for lower-income people.

**Note: unless otherwise specified, $f(x)$ refers to pay rate.**

##### Hypothesis 1-A: Income class and hourly/salaried pay

We considered that there should be a differentiation between the pay rates of individuals who are paid hourly and those who receive a salary, with salaried employees earning more. The relationship between income class and pay rate should stay the same.

$$f(x) = \beta_{0} + \beta_{middle-income}X_{middle-income} + \beta_{upper-income}X_{upper-income} +\beta_{salary}X_{salary} +\epsilon$$

Where $\beta_{0}$ represents lower-class individuals who are paid hourly.

##### Hypothesis 1-B: Income class and hours worked/week

I hypothesize that the income_class predictors will remain constant (the specific number may change, but the relationship and ability to explain Y will not). My prediction for the interaction terms is as follows: the slope of lower class will be the smallest, the slope of middle class will be slightly steeper, and the slope of upper class will be the steepest. As hours worked/week increase, pay rate should generally increase in a positive direction.

$$f(x) = \beta_{0} + \beta_{middle-income}X_{middle-income} + \beta_{upper-income}X_{upper-income} +\beta_{hours}X_{hours}+ \beta_{middle-income}X_{middle-income}(\beta_{hours}X_{hours})+ \beta_{upper-income}X_{upper-income}(\beta_{hours}X_{hours}) +\epsilon $$

Where $\beta_{0}$ represents lower-class individuals who work 0 hours per week.

### Hypothesis 2: the education hypothesis

Previous literature indicates that higher levels of academic achievement are correlated to superior job outcomes vs. lower levels of education, e.g., higher salary. (Citation needed) If higher levels of education are correlated with higher salaries, then it should stand to reason that higher levels of education should be correlated with higher pay rates, and lower levels of education are correlated with lower pay rates.

The CPS ASEC identifies 16 levels of education, from below grade school to doctoral degree. Out of mercy to the reader, the 16 levels of education are condensed in the functional form.

$$ f(x) = \beta_{0} + \beta_{grades.1.through.4}X_{grades.1.through.4}...\beta_{doctorate}X_{doctorate} + \epsilon $$

Where $\beta_{0}$ represents the pay rate of individuals with a sub-grade school education.

#### Hypothesis 2-A: the education:income class interaction

When education level and income class are visualised together, there is notable separation between classes and their pay rates within the same education levels. Therefore, there is likely an interaction effect between education and income class that can be used to more accurately describe the relationship of the predictor variables to pay rate. 

$$ f(x) = \beta_{0} + \beta_{lower.class}(\beta_{grades.1.through.4}X_{grades.1.through.4}...\beta_{doctorate}X_{doctorate}) + \beta_{middle.class}(\beta_{grades.1.through.4}X_{grades.1.through.4}...\beta_{doctorate}X_{doctorate}) + \beta_{upper.class}(\beta_{grades.1.through.4}X_{grades.1.through.4}...\beta_{doctorate}X_{doctorate}) + \epsilon$$

Where $\beta_{0}$ represents the pay rate of lower-class individuals with a sub-grade school education.

### Hypothesis 3: the major occupation hypothesis

Education was a relatively underwhelming predictor. In fact, when class was controlled for, education itself predicted nothing about pay rate.

Therefore, it was decided to predict pay rate on the basis of a new variable, previously not introduced to the data (though part of the CPS ASEC): major occupation category. Major occupation categories represent broad job classes, from the traditionally lower-class service jobs to the traditionally upper middle-class/upper-class management, business, and finance jobs (represented as "finance" for brevity). I held that major occupations, being more directly related to jobs, would naturally be more robust and accurate indicators to estimate pay rate with.

It was hypothesized that the traditionally lower-class jobs would earn smaller pay rates, and that pay rates for jobs would increase as the class of the job increased, with those in the "professional" and management, business, and finance category earning the most.

$$ f(x) = \beta_{0} + \beta_{service}X_{service}...\beta_{finance}X_{finance} + \epsilon$$

Where $\beta_{0}$ represents the estimated pay rate of an individual in the farming, fishing, and forestry industry.

#### Hypothesis 3-A: major occupation and income class

Due to temporal constraints, further investigation into the likely interaction of major occupation and income class was halted. A compromise solution was taken, whereby major occupation and income class were additively modeled. This would help determine whether income class accounted for variance in Y substantially more than major occupation category, or whether both variables were robust indicators of pay rate.

The hypothesis is that, as major occupation increases, pay rate increases. Concurrently, as income class increases, pay rate increases. Assuming both variables are robust predictors, each should substantially explain variance in Y.

$$ f(x) = \beta_{0} + \beta_{service}X_{service}...\beta_{finance}X_{finance} + \beta_{middle.class}X_{middle.class} + \beta_{upper.class}X_{upper.class} + \epsilon$$

Where $\beta_{0}$ represents the estimated pay rate of a lower-class individual in the farming, fishing, and forestry industry.

#### Hypothesis 3-B: major occupation and union coverage

There is a wealth of research indicating that being a unionized worker is correlated to higher salaries vs. non-union workers. (Citation needed) To support and identify this trend in our data, the same directionality for major occupation category was hypothesized. The interaction of union membership and major occupation category was modeled. The hypothesis is that people with no union coverage would generally have the smallest pay rate, followed by people with union coverage, and then union members would have the largest pay rates.

$$ f(x) = \beta_{0} + \beta_{no.union.coverage}X_{no.union.coverage}(\beta_{service}X_{service}...\beta_{finance}X_{finance}) + \beta_{union.coverage}X_{union.coverage}(\beta_{service}X_{service}...\beta_{finance}X_{finance}) + \beta_{union.membership}X_{union.membership}(\beta_{service}X_{service}...\beta_{finance}X_{finance}) + \epsilon$$

#### Hypothesis 3-C: subset selection model

Subset selection was used on a predetermined set of variables of interest, to determine the likely best model and reduce dimensionality. The best model kept certain levels of variables, and excluded others.

It was hypothesized that the effect of lower and upper class membership on pay rate would remain consistent. It was hypothesized that the difference between a college bachelor's and doctorate would be significant, and that college master's would have a higher pay rate than college bachelor's. It was hypothesized that professional school degrees would earn the highest pay rate.

It was hypothesized that females would make significantly less than males, and that salaried workers would make significantly less than hourly workers.

$$f(x) = \beta_0 + \beta_{upper.class}X_{upper.class} + \beta_{education.bachelor}X_{education.bachelor} + \beta_{education.masters}X_{education.masters} + \beta_{education.prof}X_{education.prof} + \beta_{education.doctorate}X_{education.doctorate} + \beta_{salary}X_{salary} + \beta_{female}X_{female} + \epsilon$$

Where $\beta_0$ represents a lower-class male with a college bachelor's who is paid hourly.

## Union members recode

### Initial Code
`A_UNMEM`: 0:NIU, 1:Yes, 2:No.
### Recode
`A_UNMEM`: 0:No, 1:Yes
### Reasoning
 The union membership variable in the above correlation table is negatively correlated with income. That makes no sense, as we know that "Workers with union representation enjoy a significant pay premium compared to non-union workers." (United States Department of Labor). The answer lies in the codes. If yes is coded to a lesser value than no, then the correlation appears to be negative. However, when those codes are flipped, the reverse is true. Both are equally representative of reality. Crucially, the latter method allows us to leverage human heuristics instead of forcing human readers to pause and consider the coding.

```r
# FILTER: UNION MEMBERS VS. NON-UNION MEMBERS
union_df_b   <- pp.job_qual  %>% 
    filter(A_UNMEM != 0) #remove NIU (children, armed forces, just in case)

dim(union_df_b) # the dim does not change once the NIU value is removed, so we do not lose any of our data
                # by applying this filter.

union_df   <- pp.job_qual  %>% 
    select(A_GRSWK, A_UNMEM) %>%
    filter(A_UNMEM != 0)
```

```r
# Recode take 1
union_df  <- union_df  %>% 
    mutate(A_UNMEM = replace(A_UNMEM, A_UNMEM == 2, 0)) #flip position of "No" relative to "Yes." Relationship otherwise held constant.

head(union_df)

# Recode boxplot
gg  <- ggplot(data = union_df, aes(group = A_UNMEM, A_GRSWK))
gg + geom_boxplot(aes(fill = A_UNMEM))
```

## Introducing the economic class factor

Household income can be used as a measure of economic class. The economic classes are defined as lower, middle, and upper. The lower class makes under 2/3 of the median income, the middle class makes from 2/3 to 2 times the median income, and the upper class makes 2 times the median income. 

This method of classification was taken from Pew Research Center analysis of the 2021 CPS, which used the University of Minnesota's Integrated Public Use Microdata Series (IPUMS). (Pew Research Center, 2022) IPUMS assigns universal codes to data over the years, which addresses the unweildly nature of across-year comparisons using the standard CPS data.

Since this study is concerned with predictors of job outcome, the classification of economic class is included in order to determine the degree of impact of economic class on job outcomes.

In order to consider household income, the datasets for individuals and households must be merged, so that we can assess the impact of household income on individuals. By doing this, we are able to make relatively more holistic judgements.

It is relevant to note that household income is a measure of the prior year. Ideally, we would want to trace housheold income back across sevral years in order to have a more accurate predictive and explanatory capability. Such an analysis is beyond the scope of this study, and is already present in the literature. The literature notes that US household income distribution has not changed significantly over the years. (Pew Research Center, 2022)

### Adjusted Income

Household income does not account for income per people the income supports. To correct for this, adjusted income can be calculated as the total household income divided by the number of persons in the household times 1/2, following other researchers. (Pew Research Center, 2022)

$$\text{adjusted household income} = \text{household income}/\text{number of household members}^{0.5}$$

As you can see in the histogram above, pay rate has a strong right skew. In other words, it follows a Gamma ($\Gamma$) distribution, which will have to be replicated in our model.

To simulate data according to a Gamma distribution, 2 parameters must be known: 
1. Shape, $\alpha$ 
    - (how right-skewed is the data?)
2. Rate, $\lambda$
    - $\lambda = \alpha/\mu(x)$, where x is a variable that follows a Gamma distribution
    
The Gamma distribution takes the following form:
$x$ ~ $\Gamma(\alpha, \lambda)$.

### Obtaining Shape
Shape can be obtained by running an inercept-only model for pay rate.

R's glm function, given that we specify that the response variable follows a Gamma distribution, returns a dispersion coefficient, which equals the shape coefficient $\alpha$ for a Gamma-distributed variable $x$.

### Defining the Simulation within a Function

```r
simulate_wages  <- function(n_subjects, beta_0, beta_1, beta_2, shape_gamma, rate_gamma, linc_prop, minc_prop, uinc_prop){
    
    y = rep(NULL, n_subjects)
    
    # Generate a vector of size n with 3 categories: lower, middle, and upper, where the probability
    # of each category occuring is given by the probability of the respective category's presence in the census data.
    income_class  <- sample(c("lower", "middle", "upper"), size = n_subjects, replace = T, prob = c(linc_prop, minc_prop, uinc_prop))
    income_class  <- as.factor(income_class)       
    
    count = 0
    for (i in income_class){
    count  <- count + 1
    if (i == "lower"){
        y[count] = beta_0 + rgamma(1, shape = shape_gamma, rate = rate_gamma)
    }
    else if (i == "middle"){
        y[count] = beta_0 + beta_1 + rgamma(1, shape = shape_gamma, rate = rate_gamma)
    }
    else {
        y[count] = beta_0 + beta_2 + rgamma(1, shape = shape_gamma, rate = rate_gamma)
    }
}
    return(data.frame(y, income_class))
}
```

The first code cell shows the structure of the simulated data, where y = pay rate and income_class represents itself.

The second shows how, despite the difference in estimate, the t and p-values for both simulations are very close to each other. The `run_analysis` function, defined in the code cell below, takes the output of the `simulate_wages` function and returns the estimated mean, standard error, and p-value for middle-class people and upper-class people, sequentially. The first 3 items in the vector correspond to the middle class, and the last 3 correspond to the upper class.

```r
run_analysis  <- function(data) {
    h0  <- glm(data$y ~ 1, family = Gamma("identity"))
    ha  <- glm(data$y~ data$income_class, family = Gamma("identity") )
    smry_ha  <- summary(ha)
    
   # print(smry_ha$coefficients)
    
    #est_base  <-  smry_ha$coefficients[, 1, drop = FALSE][1]
    est_m  <- smry_ha$coefficients[, 1, drop = FALSE][2]
    est_u  <- smry_ha$coefficients[, 1, drop = FALSE][3]
    
    # standard errors
    se_m  <- smry_ha$coefficients[, 2, drop = FALSE][2] #for middle class
    se_u  <- smry_ha$coefficients[, 2, drop = FALSE][3] #for upper class
    
    m_aic  <- AIC(h0, ha)$AIC
    #print(m_aic)
    statistic  <- diff(m_aic)
    #paste(m_aic)
    p_m  <- smry_ha$coefficients[, 4, drop = FALSE][2]
    p_u  <- smry_ha$coefficients[, 4, drop = FALSE][3]
    
    return(c(est_m, se_m, p_m, est_u, se_u, p_u))
}
```

## Full Monte Carlo Analysis

### Function

The `repeat_analysis` function takes the same inputs as the `simulate_wages` function, in addition to the number of simulations and alpha.

```r
repeat_analysis  <- function(n_simulations, alpha, n_subjects, beta_0, beta_1, beta_2, 
                             shape, rate, prop_0, prop_1, prop_2) {
    
    simouts <- matrix(rep(NA,6*n_simulations),nrow=n_simulations) # empty vector to store outputs from each simulation
    
    #loop for iterating simulation
    for (i in 1:n_simulations) {
        data  <- simulate_wages(n_subjects, beta_0, beta_1, beta_2, shape, rate, prop_0, prop_1, prop_2)
        simouts[i,] <- run_analysis(data) # save the analysis outputs for this sim.
    }
    
    
    #print(mean(simouts[,3]))
    power_m  <- mean(simouts[,3] <= alpha) # of analyses where p < alpha; aka average power
    power_u  <- mean(simouts[,6] <= alpha)

    theta_bias_b1 <- ( mean(simouts[,1]) - beta_1 ) / (beta_1)
    theta_bias_b2 <- ( mean(simouts[,4]) - beta_2 ) / (beta_2)
    
    sigma_bias_b1 <- (mean(simouts[,2]) - sd(simouts[,1])) / sd(simouts[,1])
    sigma_bias_b2 <- (mean(simouts[,5]) - sd(simouts[,4])) / sd(simouts[,4])
    
    return(list(power_b1 = power_m, theta_b1  = theta_bias_b1, sigma_b1 = sigma_bias_b1,
               power_b2 = power_u, theta_b2 = theta_bias_b2, sigma_b2 = sigma_bias_b2))
}
```

```r
# define function
estimates_identity  <- function(matrix) {
    output  <- c()
    
    intercept  <- matrix[1,1]
    
    for (i in 1:length(matrix[,1])) {
        if (i ==1) {
            output[i]  <- intercept
        }
        else {
            output[i]  <- matrix[i,1] + intercept
        }
    }
    return(output)
} 
```

## Limitations

For models to be accurate, they data must meet the underlying assumptions of the model. The generalized linear model is a form of linear model, so these models must meet the underlying assumptions.

1. Linearity assumption | To my knowledge, all models describe a linear relationship between the predictors and response variable. The functional form of the models describes a linear relationship, and therefore should meet this assumption.

2. Y is Gamma-distributed | Often, one assumes a normal distribution. However, `pay_rate` is clearly right-skewed. Thus, all models predict pay rate assuming a Gamma distribution.

3. No colinearity between features | To my knowledge, there should be no colinearity between features in the models. However, checks were not conducted to ensure this assumption was met. Therefore, the model results should be treated as invalid until proven otherwise.

4. f(x) is stationary | The rare quantitative variables included in analysis are stationary, in that for every point $x_i$ there is only one point $y_i$. Regression using qualitative predictors solves for this problem mathematically by using separate functions to calculate estimates for any given level of a variable, resulting in multiple points describing the relationship between different levels of a variable $v$ and its impact on f(x). It should be noted that these estimates represent single points, typically the mean of the level of a variable.

Interpretation of results was conducted assuming that these assumptions have been met. However, that does not make it true. I wholeheartedly encourage unrestricted reproduction of these analyses, whether it be with the intent to prove or disprove them. I especially encourage criticisms of the methodology in this study so it can be rectified in future studies. It is a critical part of continued learning, and I should feel so honored to have someone take the time to point out issues with the core beliefs of this study, its methodology, and the interpretation of its results.

This is my first foray into a major statistical analysis, and was largely conducted by myself. As such, I am certain there are a plethora of errors I have overlooked. The study results should be interpreted in light of this fact.

I am aware that assigning weights to different levels of variables allows for more precise estimates. I have not learned how to apply them, however. All studies are encouraged to avoid this titanic error in model construction and data architecture. To grossly oversimplify The Great British Class Survey methodology, it used weights to lessen the weight of middle and upper-class individuals (who were overrepresented vs. lower-class individuals). Otherwise, the sheer n of upper-class individuals vs. other levels of the variable would lead to data inaccuracy. Therefore, the application of weights is crucial to ensure all levels of variables are comparable and have similar abilities to explain variance in Y. 

For reason of incompetence and inexperience, weights were regrettably not used. I strongly urge future studies to avoid this massive methodological error.

I would also caution against any strong belief in the accuracy of these coefficients in describing pay rates relative to each other. By intuition and by virtue of their similarity to previously-measured trends (Citation(s) needed), the measures appear vaguely accurate. But they are not nearly accurate enough to constitute a legitimate piece of data science, only an attempt at one.

Considering the replication crisis in the Psychology field (Citation needed), which I should assume is present in other social sciences (such as the nature of this work), the inaccuracy of these models represents another contribution to inaccurate data produced in the field of data science. This is massively disappointing.

Fortunately, the entire code and methodology of this data is readily present in this Jupyter Notebook. #### Income Class

Lower-class status appears to correlated with the lowest wages. The degree to which lower class explains wages is difficult to determine, on account of the low n compared to other levels of income class and its existence as the baseline of each model.

Nevertheless, in terms of differentiation from the upper and middle class, the effect of lower class, in every model, was (with one exception, the income class + `paid_hourly` model) the lowest.

Middle class proved a relatively unstable predictor when modeled alongside other variables, but was generally associated with the second-lowest levels of pay rates when modeled by itself and in interactions with education and major occupation.

Upper-class status was one of the most robust predictors, with models consistently detecting a high degree of significant effect on pay rate.

The model estimates reflect high inequality between the lower and middle class, with the middle class earning (without exception) higher wages, though the impact of middle class on Y was not always significant.

The model estimates reflect very high inequality between the upper class and the lower + middle classes. Upper-class individuals always had the highest pay rates versus these two classes.

Extrapolating, this demonstrates the need for class solidarity between the lower and middle classes against the upper class, who earn disproportionately more than them. Although pay rate is mathematically determined, in that individuals with a higher salary will always have a higher pay rate compared to individuals with lower salaries, what pay rate means on a human level is more than that.

If pay rate is how much your labor is valued, and your labor is inherently tied to your identity, it is absurd that such a high degree of disparity exist, particularly between the upper class and everyone else. One hour is one hour. Some occupations and businesses earn more money by virtue of their market profitability, and there is nothing wrong with that. What is fundamentally disturbing is that some individuals' labor should be valued so high above others', when both are human beings.

With wages reflecting income, the lower labor value of lower and middle-class individuals represent that they likely have less spending money, requiring them to endure economic hardships that are alien to individuals with more spending money.

In a society where all individuals are truly treated as equals, no individual should have to endure more economic hardships on account of their class. Income ought to be redistributed to the lower and middle classes at cost to the upper class, who already have more than they need.

A society is fundamentally a collective, and no degree of individualist philosophy can change that. Major Occupation estimates showed considerable discrepancies between major occupation and their level of pay. Major occupation appeared to be a substiantially more robust predictor than education_level, as the strength of most correlations in terms of t- and p- value was reduced but not negated by being modeled alongside income class (unlike education).

I extrapolate that future studies may find fruitful answers in terms of the relationship of major occupation to class.
